### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
pd.set_option('display.max_rows', None)
df.info(verbose=True, show_counts=True)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
EdLevel_counts = df['EdLevel'].value_counts()
print(EdLevel_counts)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
ed_level_map = {
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'High School',
    'Some college/university study without earning a degree': 'Undergraduate',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional',
    'Associate degree (A.A., A.S., etc.)': 'Associate'
}

df['EdLevel'] = df['EdLevel'].replace(ed_level_map) 
print(df['EdLevel'].unique())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here

df_one_hot = pd.get_dummies(df, columns=['Employment'], dtype=int)
print(df_one_hot.head())
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values.head(15))
print(df.shape)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

median_value_CCY = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_value_CCY)

print(f"Empty rows is: {df['ConvertedCompYearly'].isnull().sum()}")
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
freq_value_RW = df['RemoteWork'].mode()[0]
print(freq_value_RW)
df['RemoteWork'] = df['RemoteWork'].fillna(freq_value_RW)
print(f"Empty cells in RW column: {df['RemoteWork'].isnull().sum()}")
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
min_val = df['ConvertedCompYearly'].min()
max_val = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly'] = (df['ConvertedCompYearly'] - min_val) / (max_val - min_val)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly_log'] = np.log1p(df['ConvertedCompYearly'])

print(df[['ConvertedCompYearly', 'ConvertedCompYearly_log']].head())

print("Skewness voor transformatie:", df['ConvertedCompYearly'].skew())
print("Skewness na transformatie:", df['ConvertedCompYearly_log'].skew())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
#To check the daata
print(df['YearsCodePro'].head(20))
print(df['YearsCodePro'].isnull().sum())

#To transform the data from a str to a float
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
print(df['YearsCodePro'].dtype)

#To create segmentation based on the years of experience
bins = [-1, 2, 5, 10, np.inf]
labels = ['Junior', 'Intermediate', 'Senior', 'Expert']
df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels) 


print(df['ExperienceLevel'].value_counts())
print(df['ExperienceLevel'].head(10))
```
