### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_count = df.isnull().sum()
missing_info = pd.DataFrame({
    'Amount': missing_count
}).sort_values(by='Amount', ascending=False)

pd.set_option('display.max_rows', None)

print(missing_info)

```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
plt.figure(figsize=(16,25)) 

sns.heatmap(missing_info, cbar=True, yticklabels=True, xticklabels=False, cmap='binary')

plt.title('Heatmap missing values dataset)')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here

#in the data employment column isn't missing any data therefore using DevType
missing_count_Emp = df['DevType'].isnull().sum()
print(missing_count_Emp)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
counts = df['DevType'].value_counts()

top_value_DevType = counts.index[0]
frequency_DevType = counts.iloc[0]

print(f"profession '{top_value_DevType}' is selected {frequency_DevType} as DevType.")
```

##### Task 6: Impute missing values in the `DevType` column with the most frequent value.



```python
## Write your code here
df['DevType'] = df['DevType'].fillna(top_value_DevType)
print(f"Amount of missing values at the moment: {df['DevType'].isnull().sum()}")
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plt.figure(figsize=(12, 6))
sns.countplot(y='DevType', data=df, order=df['DevType'].value_counts().iloc[:15].index, 
              color='salmon', label='Na Imputatie')
plt.title('Most popular DevType count')
plt.legend()
plt.show()
```
